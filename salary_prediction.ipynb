# Categorising countries

```python
df['Country'].value_counts()
```

```python
def shorten_categories(categories, cutoff):
    categor_map = {}
    for i in range(len(categories)):
        if categories.values[i] >= cutoff:
            categor_map[categories.index[i]] = categories.index[i]
        else:
            categor_map[categories.index[i]] = 'Other'
    return categor_map
```

```python
countries = shorten_categories(df.Country.value_counts(), 300)
```

```python
df['Country'] =  df['Country'].map(countries)
df.Country.value_counts()
```

```python
fig, ax = plt.subplots(1,1, figsize=(15,8))
df.boxplot('Salary', 'Country', ax=ax)
plt.xticks(rotation=90)
plt.show()
```

# Data cleaning

```python
df['YearsCodePro'].unique()
```

```python
def clean_experience(x):
    if x == 'More than 50 years':
        return 50
    if x == 'Less than 1 year':
        return 0.5
    return float(x)
```

```python
df['YearsCodePro'] = df['YearsCodePro'].apply(clean_experience)
```

```python
df['EdLevel'].unique()
```

```python
def clean_education(x):
    if 'Bachelor’s degree' in x:
        return 'Bachelor’s degree'
    if 'Master’s degree' in x:
        return 'Master’s degree'
    if 'Professional degree' in x:
        return 'Professional degree'
    return 'Less than Bachelor’s'

df['EdLevel'] = df['EdLevel'].apply(clean_education)
```

```python
df['EdLevel'].unique()
```

# Fine tuning a model

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [None, 2, 4, 6, 8, 10, 12, 15, 20, 25],
    'min_samples_split': [2, 3, 4, 5, 10, 15, 20],
    'min_samples_leaf': [1, 2, 4, 6, 8, 10]
}

regressor = DecisionTreeRegressor(random_state=13)
gs = GridSearchCV(regressor, param_grid, scoring='neg_mean_squared_error', cv=5, n_jobs=-1)
gs.fit(X, y)
```

```python
regressor = gs.best_estimator_

regressor.fit(X, y.values)
y_pred = regressor.predict(X)
error = np.sqrt(mean_squared_error(y, y_pred))
print("${:,.02f}".format(error))
```

```python
X_test = test_set.drop('Salary', axis=1)
y_test = test_set['Salary']
```

```python
best_regressor = gs.best_estimator_
best_regressor.fit(X, y)
y_pred_train = best_regressor.predict(X)
y_pred_test = best_regressor.predict(X_test)
error_train = np.sqrt(mean_squared_error(y, y_pred_train))
error_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
print("Tuned model train error: ${:,.02f}".format(error_train))
print("Tuned model test error: ${:,.02f}".format(error_test))
```
